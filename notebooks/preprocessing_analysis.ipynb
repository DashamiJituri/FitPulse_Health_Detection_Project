```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data_clean/cleaned_fitness_data.csv", parse_dates=["timestamp"])

df.describe()

plt.figure(figsize=(10,4))
plt.plot(df["timestamp"], df["heart_rate"])
plt.title("Heart Rate Trend")
plt.show()

plt.figure(figsize=(10,4))
plt.plot(df["timestamp"], df["steps"])
plt.title("Steps Trend")
plt.show()
```
